```python
import os
import shutil
from sklearn.model_selection import train_test_split

# train
train_img_path = './ISIC2017/ISIC-2017_Training_Data/'
train_mask_path = './ISIC2017/ISIC-2017_Training_Part1_GroundTruth/'

train_img_files = os.listdir(train_img_path)

train_images = [f for f in train_img_files if 'superpixels' not in f]
train_masks = os.listdir(train_mask_path)

# valid
valid_img_path = './ISIC2017/ISIC-2017_Validation_Data/'
valid_mask_path = './ISIC2017/ISIC-2017_Validation_Part1_GroundTruth/'

valid_img_files = os.listdir(valid_img_path)

valid_images = [f for f in valid_img_files if 'superpixels' not in f]
valid_masks = os.listdir(valid_mask_path)

# test
test_img_path = './ISIC2017/ISIC-2017_Training_Data/'
test_mask_path = './ISIC2017/ISIC-2017_Training_Part1_GroundTruth/'

test_img_files = os.listdir(test_img_path)

test_images = [f for f in test_img_files if 'superpixels' not in f]
test_masks = os.listdir(test_mask_path)



```

```python
base_path = './ISIC2017/'
#      
for folder in ['train', 'valid', 'test']:
    os.makedirs(f'{base_path}/{folder}/images', exist_ok=True)
    os.makedirs(f'{base_path}/{folder}/masks', exist_ok=True)

#   
def move_files(files, source, destination):
    for file in files:
        shutil.copy(os.path.join(source, file), os.path.join(destination, file))


#   
move_files(train_images, train_img_path, './ISIC2017/train/images')
move_files(valid_images, valid_img_path, './ISIC2017/valid/images')
move_files(test_images, test_img_path, './ISIC2017/test/images')

move_files(train_masks, train_mask_path, './ISIC2017/train/masks')
move_files(valid_masks, valid_mask_path, './ISIC2017/valid/masks')
move_files(test_masks, test_mask_path, './ISIC2017/test/masks')
```
